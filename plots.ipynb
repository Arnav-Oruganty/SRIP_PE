```python
import json
import matplotlib.pyplot as plt
from matplotlib.ticker import LogFormatterMathtext

# Load data
with open("results.txt") as f:
    data = json.load(f)

nmse_vs_snr = data["nmse_vs_snr"]
nmse_vs_b = data["nmse_vs_b"]

SNR_vals = [0, 2, 4, 6, 8, 10]
B_vals = sorted(int(k) for k in nmse_vs_b.keys())

# Plot style
plt.rcParams.update({
    "font.size": 12,
    "axes.labelweight": "bold",
    "axes.linewidth": 1.5,
})

# ---------- NMSE vs SNR ----------
for N in ["2", "3"]:
    fig, ax = plt.subplots(figsize=(8,5))

    for method, marker in zip(["Random","SA","PSO","LA-MCTS-PO"], ["o","o","o","o"]):
        ax.plot(
            SNR_vals,
            nmse_vs_snr[N][method],
            marker=marker,
            linestyle="--",
            linewidth=3.0,
            markersize=7,
            label=method
        )

    ax.set_xlabel("SNR (dB)", fontweight="bold")
    ax.set_ylabel("NMSE", fontweight="bold")
    ax.set_yscale("log")

    # Only show 10^0, 10^-1, 10^-2
    ax.set_yticks([1e0, 1e-1, 1e-2])
    ax.yaxis.set_major_formatter(LogFormatterMathtext())

    # Optional: remove minor tick labels
    ax.yaxis.set_minor_formatter(plt.NullFormatter())

    ax.grid(True, which="both", linestyle="--", linewidth=0.8)

    legend = ax.legend(frameon=True, fontsize=10)
    for text in legend.get_texts():
        text.set_fontweight("bold")

    fig.tight_layout()
    fig.savefig(f"NMSE_vs_SNR_N{N}.pdf", dpi=300)
    fig.savefig(f"NMSE_vs_SNR_N{N}.eps", format="eps")
    plt.close(fig)

# ---------- NMSE vs B ----------
fig, ax = plt.subplots(figsize=(8,5))

for method, marker in zip(["Random","SA","PSO","LA-MCTS-PO"], ["o","o","o","o"]):
    ax.plot(
        B_vals,
        [nmse_vs_b[str(B)][method][0] for B in B_vals],
        marker=marker,
        linestyle="--",
        linewidth=3.0,
        markersize=7,
        label=method
    )

ax.set_xlabel("Number of RIS elements (B)", fontweight="bold")
ax.set_ylabel("NMSE", fontweight="bold")
ax.set_yscale("log")

# Only show 10^0, 10^-1, 10^-2
ax.set_yticks([1e0, 1e-1, 1e-2])
ax.yaxis.set_major_formatter(LogFormatterMathtext())

# Optional: remove minor tick labels
ax.yaxis.set_minor_formatter(plt.NullFormatter())

ax.grid(True, which="both", linestyle="--", linewidth=0.8)

ax.set_xticks(B_vals)
legend = ax.legend(frameon=True, fontsize=10)
for text in legend.get_texts():
    text.set_fontweight("bold")

fig.tight_layout()
fig.savefig("NMSE_vs_B.pdf", dpi=300)
fig.savefig("NMSE_vs_B.eps", format="eps")
plt.close(fig)
```
