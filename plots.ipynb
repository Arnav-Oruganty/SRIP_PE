```python
import json
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import LogFormatterMathtext
from scipy.io import savemat

# Load data
with open("results.txt") as f:
    data = json.load(f)

nmse_vs_snr = data["nmse_vs_snr"]
nmse_vs_b = data["nmse_vs_b"]

SNR_vals = np.array([0, 2, 4, 6, 8, 10])
B_vals = np.array(sorted(int(k) for k in nmse_vs_b.keys()))

# Style
plt.rcParams.update({
    "font.size": 12,
    "axes.labelweight": "bold",
    "axes.linewidth": 1.5,
})

# Marker styles (FIXED)
markers = {
    "Random": "o",
    "SA": "s",
    "PSO": "^",
    "LA-MCTS-PO": "D"
}

# ======================================================
# NMSE vs SNR
# ======================================================
for N in ["2", "3"]:
    fig, ax = plt.subplots(figsize=(8,5))

    mat_dict = {"SNR": SNR_vals}

    for method in ["Random","SA","PSO","LA-MCTS-PO"]:
        values = np.array(nmse_vs_snr[N][method])

        # Rename for plot
        label = "LA-MCTS (Proposed)" if method == "LA-MCTS-PO" else method

        ax.plot(
            SNR_vals,
            values,
            marker=markers[method],
            linestyle="-",   # FIXED (solid)
            linewidth=3.0,
            markersize=7,
            label=label
        )

        # Save for MATLAB
        mat_dict[label.replace(" ", "_")] = values

    ax.set_xlabel("SNR (dB)", fontweight="bold")
    ax.set_ylabel("NMSE", fontweight="bold")
    ax.set_yscale("log")

    ax.set_yticks([1e0, 1e-1, 1e-2])
    ax.yaxis.set_major_formatter(LogFormatterMathtext())
    ax.yaxis.set_minor_formatter(plt.NullFormatter())

    ax.grid(True, which="both", linestyle="--", linewidth=0.8)

    legend = ax.legend(frameon=True, fontsize=10)
    for text in legend.get_texts():
        text.set_fontweight("bold")

    fig.tight_layout()

    # Save files
    fig.savefig(f"NMSE_vs_SNR_N{N}.pdf", dpi=300)
    fig.savefig(f"NMSE_vs_SNR_N{N}.eps", format="eps")

    # Save MAT file
    savemat(f"NMSE_vs_SNR_N{N}.mat", mat_dict)

    plt.close(fig)

# ======================================================
# NMSE vs B
# ======================================================
fig, ax = plt.subplots(figsize=(8,5))

mat_dict = {"B": B_vals}

for method in ["Random","SA","PSO","LA-MCTS-PO"]:
    values = np.array([nmse_vs_b[str(B)][method][0] for B in B_vals])

    label = "LA-MCTS (Proposed)" if method == "LA-MCTS-PO" else method

    ax.plot(
        B_vals,
        values,
        marker=markers[method],
        linestyle="-",   # FIXED (solid)
        linewidth=3.0,
        markersize=7,
        label=label
    )

    mat_dict[label.replace(" ", "_")] = values

ax.set_xlabel("Number of Pilot Frames (B)", fontweight="bold")
ax.set_ylabel("NMSE", fontweight="bold")
ax.set_yscale("log")

ax.set_yticks([1e0, 1e-1, 1e-2])
ax.yaxis.set_major_formatter(LogFormatterMathtext())
ax.yaxis.set_minor_formatter(plt.NullFormatter())

ax.grid(True, which="both", linestyle="--", linewidth=0.8)

ax.set_xticks(B_vals)

legend = ax.legend(frameon=True, fontsize=10)
for text in legend.get_texts():
    text.set_fontweight("bold")

fig.tight_layout()

# Save files
fig.savefig("NMSE_vs_B.pdf", dpi=300)
fig.savefig("NMSE_vs_B.eps", format="eps")

# Save MAT file
savemat("NMSE_vs_B.mat", mat_dict)

plt.close(fig)
```
